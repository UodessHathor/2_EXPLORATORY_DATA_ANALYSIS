##### *Creation of a categorical dataframe which will be used for categorical variables visualizations*

```python
import matplotlib.pyplot as plt
from matplotlib import cm
from math import log10

# -------------- REEVALUATION ATTRIBUTES (average scores) : one day after -------------------------------------------------
oda_7_labels_female = ["Attractive", "Fun", "Sincere", "Intelligent", "Shared Interests", "Ambitious"]
oda_7_data_female = [28.68, 17.64, 17.36, 16.89, 10.57, 8.87]
oda_7_labels_male = ["Attractive", "Fun", "Intelligent", "Shared Interests", "Sincere", "Ambitious"]
oda_7_data_male = [33.21, 29.74, 15.45, 9.97, 6.89, 4.74]


# ------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------
n = len(oda_7_data_male)
k = 10 ** int(log10(max(oda_7_data_male)))
m = k * (1 + max(oda_7_data_male) // k)
r = 1.5
w = r / n

# ------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

colors_females = [cm.PuOr(i / n) for i in range(n)]
colors_males = [cm.cividis(i / n) for i in range(n)]

fig, axs = plt.subplots(1, 2, figsize=(16,4))
axs[0].axis("equal")
axs[1].axis("equal")

# BEFORE GRAPH -----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------
for i in range(n): # women 
    innerring, _ = axs[0].pie([m - oda_7_data_female[i], oda_7_data_female[i]], radius = r - i * w, startangle = 90, labels = ["", oda_7_labels_female[i]], labeldistance = 1 - 1 / (1.5 * (n - i)), textprops = {"alpha": 0}, colors = ["white", colors_females[i]])
    plt.setp(innerring, width = w, edgecolor = "white")

for i in range(n): # men
    innerring2, _ = axs[1].pie([m - oda_7_data_male[i], oda_7_data_male[i]], radius = r - i * w, startangle = 90, labels = ["", oda_7_labels_male[i]], labeldistance = 1 - 1 / (1.5 * (n - i)), textprops = {"alpha": 0}, colors = ["white", colors_males[i]])
    plt.setp(innerring2, width = w, edgecolor = "white")


axs[0].text(0, 2, 'Think back : what attributes best reflects the importance of these in your decisions ?', fontsize = 20) 
axs[0].text(-2.3, -0.2, 'Women', fontsize = 14)
axs[1].text(0, -0.1, 'Men', fontsize = 14)
axs[1].text(2, 0, 'Selection of those with yes match and second date', fontsize = 14)
axs[0].legend(loc='upper left', fontsize=8)
axs[1].legend(loc='upper left', fontsize=8)

plt.show()
```
